```python
import cv2
import matplotlib.pyplot as plt

# Etapa 1: Baixar o classificador Haar Cascade
cascade_url = "https://github.com/opencv/opencv/raw/master/data/haarcascades/haarcascade_frontalface_default.xml"
cascade_path = "haarcascade_frontalface_default.xml"

import urllib.request
urllib.request.urlretrieve(cascade_url, cascade_path)

# Etapa 2: Carregar a imagem
imagem_nome = "especialista-ou-generalista (2).jpg"
img = cv2.imread(imagem_nome)

# Etapa 3: Converter para escala de cinza
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Etapa 4: Carregar o classificador e detectar rostos
face_cascade = cv2.CascadeClassifier(cascade_path)
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

# Etapa 5: Desenhar retângulos nos rostos detectados
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)

# Etapa 6: Exibir a imagem com os rostos destacados
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.title(f"{len(faces)} rosto(s) detectado(s)")
plt.show()
```
